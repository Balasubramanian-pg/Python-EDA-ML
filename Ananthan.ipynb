```python
import pandas as pd
from IPython.display import display

# Read the data
df = pd.read_excel(r"C:\Users\balas\Downloads\Anandhan Responses.xlsx")  # Use read_excel for Excel files

# Select categorical columns for one-hot encoding
categorical_columns = ['Age', 'concept_fam', 'prev_engag', 'freq', 'engage', 'games', 'interaction', 'purchase_response', 'engage_factors', 'privacy', 'rewards_incentives', 'competiton_based', 'social_media_platform', 'design_aesthetics', 'targeting', 'challenges', 'personalization', 'sharing_rate']

# Perform one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_columns)

# Display horizontally scrollable DataFrame
with pd.option_context('display.max_columns', None):
    display(df_encoded.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix_encoded = df_encoded.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix_encoded, cmap='coolwarm', annot=False, fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix (One-Hot Encoded)')
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

# Load the dataset (replace 'your_dataset.xlsx' with the actual file name)
df = pd.read_excel(r"A:\Semester 4\1.Dissertation\Analysis and Interpretation\Anandhan Responses.xlsx")

# Choose two categorical variables for analysis
variable1 = 'prev_engag'
variable2 = 'concept_fam'

# Create a contingency table
contingency_table = pd.crosstab(df[variable1], df[variable2])

# Perform Chi-squared test for independence
chi2, p, _, _ = chi2_contingency(contingency_table)

# Calculate Cramér's V statistic
num_rows = contingency_table.shape[0]
num_columns = contingency_table.shape[1]
cramers_v = np.sqrt(chi2 / (num_rows * (min(num_columns, num_rows) - 1)))

# Print results
print(f"Chi-squared value: {chi2}")
print(f"P-value: {p}")
print(f"Cramér's V statistic: {cramers_v}")

# Interpret results
alpha = 0.05
print("\nInterpretation:")
print(f"Significance level: {alpha}")
print(f"P-value: {p}")
if p < alpha:
    print("Reject the null hypothesis (There is a significant relationship between the variables)")
else:
    print("Fail to reject the null hypothesis (There is no significant relationship between the variables)")

# Cramér's V interpretation
print("\nCramér's V Interpretation:")
print(f"A Cramér's V value of 0 indicates no association, and 1 indicates a strong association.")
print(f"Cramér's V: {cramers_v}")
```
