```python
import pandas as pd
import pingouin as pg
from sklearn.preprocessing import LabelEncoder

# Load the dataset from an Excel file
excel_file_path = r"A:\Semester 4\1.Dissertation\Analysis and Interpretation\Ananthu.xlsx"  # Update with your file path
data = pd.read_excel(excel_file_path)

# Extract columns that you want to include in Cronbach's alpha calculation
selected_columns = data[['CRM_Familiarity', 'Exposure_CRM', 'Customer_loyalty', 'Loyalty_CRM',
                         'Most_appealing_CRM', 'Companies_obtain_info', 'Channels',
                         'Frequency', 'Privacy_concern', 'data_trust', 'Feedback',
                         'CRM_Influence_on_Loyalty', 'CRM_Comparison', 'Future_scope']]

# Convert string data to numeric values
label_encoder = LabelEncoder()
selected_columns_numeric = selected_columns.apply(label_encoder.fit_transform)

# Calculate Cronbach's alpha
alpha_result = pg.cronbach_alpha(data=selected_columns_numeric)

# Display the Cronbach's alpha value
print("Cronbach's Alpha:")
print(alpha_result[0])  # Access the first element of the tuple, which is the alpha value
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load the dataset from an Excel file
excel_file_path = r"A:\Semester 4\1.Dissertation\Analysis and Interpretation\Ananthu.xlsx"  # Update with your file path
data = pd.read_excel(excel_file_path)

# Assuming you have two categorical columns A and B, replace them with your actual column names
column_A = 'Exposure_CRM'
column_B = 'CRM_Familiarity'

# Create a contingency table
contingency_table = pd.crosstab(data[column_A], data[column_B])

# Perform the chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Display the results
print(f"Chi-square value: {chi2}")
print(f"P-value: {p}")

# Check for significance
alpha = 0.05
if p < alpha:
    print("There is a significant relationship between the two variables.")
else:
    print("There is no significant relationship between the two variables.")
```
