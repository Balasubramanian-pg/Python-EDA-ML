```python
from nanonets import NANONETSOCR
model = NANONETSOCR()
model.set_token('6a7c8b27-466b-11ef-b890-76cd27428090')
```

```python
# PDF / Image to Raw OCR Engine Output
import json
pred_json = model.convert_to_prediction(r"C:\Users\ASUS\OneDrive\Pictures\Screenshots\Screenshot 2024-03-12 152807.png")
print(json.dumps(pred_json, indent=2))
```

```python
# PDF / Image to String
string = model.convert_to_string(r"C:\Users\ASUS\OneDrive\Pictures\Screenshots\Screenshot 2024-03-12 152807.png")
print(string)
```

```python
# PDF / Image to TXT File
model.convert_to_txt(r"C:\Users\ASUS\OneDrive\Pictures\Screenshots\Screenshot 2024-03-12 152807.png", output_file_name = r"C:\Users\ASUS\OneDrive\Pictures\Output.txt")
```

```python
# PDF / Image to Boxes
# each element contains predicted word and bounding box information
# bounding box information denotes the spatial position of each word in the file
boxes = model.convert_to_boxes(r"C:\Users\ASUS\OneDrive\Pictures\Screenshots\Screenshot 2024-03-12 152807.png")
for box in boxes:
    print(box)
```

```python
# PDF / Image to CSV
# This method extracts tables from your file and prints them in a .csv file.
# NOTE : This particular function is a trial offering 1000 pages of use.
# To use this at scale, please create your own model at app.nanonets.com --> New Model --> Tables.
model.convert_to_csv(r"C:\Users\ASUS\OneDrive\Pictures\Screenshots\Screenshot 2024-03-12 152807.png", output_file_name = r"C:\Users\ASUS\OneDrive\Pictures\Book1.csv")
```

```python
# PDF / Image to Tables
# This method extracts tables from your file and returns a json object.
# NOTE : This particular function is a trial offering 1000 pages of use.
# To use this at scale, please create your own model at app.nanonets.com --> New Model --> Tables.
import json
tables_json = model.convert_to_tables(r"C:\Users\ASUS\OneDrive\Pictures\Screenshots\Screenshot 2024-03-12 152807.png")
print(json.dumps(tables_json, indent=2))
```

```python

# PDF / Image to Searchable PDF
model.convert_to_searchable_pdf(r"C:\Users\ASUS\OneDrive\Pictures\Screenshots\Screenshot 2024-03-12 152807.png", output_file_name = 'OUTPUTNAME.pdf')
```

```python
from nanonets import NANONETSOCR
import json

model = NANONETSOCR()
model.set_token('51137df8-466d-11ef-9688-e609ba4a0ccd')

try:
    response = model.extract_tables(r"C:\Users\ASUS\OneDrive\Pictures\Screenshots\Screenshot 2024-03-12 152807.png")
    print("Full API Response:")
    print(json.dumps(response, indent=2))
except Exception as e:
    print(f"An error occurred: {str(e)}")
    if hasattr(e, 'response'):
        print("Response content:")
        print(e.response.text)
```
